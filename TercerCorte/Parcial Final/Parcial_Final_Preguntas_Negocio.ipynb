## 2. Ingesta y Unificaci√≥n de Datos  
**Objetivo de negocio:** asegurar que estamos comparando peras con peras (Shein vs Temu en una misma tabla limpia).

### 2.1. Preguntas de negocio

1. **Foto grande del dataset**
   - ¬øCu√°ntas publicaciones tenemos de Shein y cu√°ntas de Temu?  
   - ¬øEsta diferencia de volumen ya nos dice algo de la estrategia de cada marca?

2. **Comparabilidad**
   - Si Shein tiene mucho m√°s contenido que Temu, ¬øpor qu√© NO podemos comparar solo el n√∫mero de likes o de posts?  
   - ¬øQu√© riesgos habr√≠a para una decisi√≥n de inversi√≥n si comparamos m√©tricas crudas sin normalizar?

3. **Estructura de la tabla unificada**
   - En el dataframe final unificado (por ejemplo, `df`), ¬øqu√© columnas consideras **clave de negocio** para comparar marcas? (ej: `brand`, `likes`, `comments`, `views`, `published_at`, etc.).  
   - Si una columna no existe en una de las dos fuentes (por ejemplo, `views` solo para algunos posts), ¬øc√≥mo afecta eso la interpretaci√≥n del desempe√±o?

4. **Representatividad**

   - Si no lo es, ¬øc√≥mo podr√≠a sesgar tus conclusiones sobre qui√©n lo est√° haciendo ‚Äúmejor‚Äù?


```python
import pandas as pd
import numpy as np
import re
from datetime import datetime

def _norm(col):
    return re.sub(r"[^a-z0-9]+", "", str(col).strip().lower())

SYNONYMS = {
    "post_id": [
        "post_id","id","videoid","video_id","awemeid","aweme_id","itemid","item_id","uri"
    ],
    "caption": [
        "caption","desc","description","text","content","title","sharetitle","videodesc"
    ],
    "published_at": [
        "published_at","publishat","publish_time","publishtime","created_at","create_time","createtime",
        "date","datetime","timestamp","time"
    ],
    "likes": [
        "likes","likescount","likecount","diggcount","statsdiggcount","heartcount"
    ],
    "comments": [
        "comments","commentscount","commentcount","statscommentcount"
    ],
    "shares": [
        "shares","sharescount","sharecount","statssharecount"
    ],
    "views": [
        "views","viewscount","viewcount","playcount","statsplaycount","videoviewcount","impressions"
    ],
    "url": [
        "url","shareurl","link","videourl","posturl"
    ],
}

def pick_column(df, candidates):
    norm_cols = {_norm(c): c for c in df.columns}
    for cand in candidates:
        key = _norm(cand)
        if key in norm_cols:
            return norm_cols[key]
    # intento adicional: columnas con puntos (ej stats.diggCount)
    for c in df.columns:
        if _norm(c) in [_norm(x) for x in candidates]:
            return c
    return None

def coerce_numeric(series):
    if series is None:
        return series
    s = series.copy()
    if s.dtype == "object":
        s = (s.astype(str)
              .str.replace(",", "", regex=False)
              .str.replace(" ", "", regex=False)
              .str.replace("\u200b", "", regex=False))
    return pd.to_numeric(s, errors="coerce")

def parse_datetime(series):
    if series is None:
        return series
    s = series.copy()

    # Si viene como n√∫mero (epoch seconds/ms)
    num = pd.to_numeric(s, errors="coerce")
    if num.notna().mean() > 0.5:
        # heur√≠stica para epoch
        v = num.dropna()
        if len(v) > 0:
            med = float(v.median())
            # ms si es > 1e12 aprox
            unit = "ms" if med > 1e12 else "s"
            dt = pd.to_datetime(num, unit=unit, errors="coerce", utc=True).dt.tz_convert(None)
            return dt

    # Si viene como string fecha
    dt = pd.to_datetime(s, errors="coerce", utc=True).dt.tz_convert(None)
    return dt

def standardize(df, brand):
    df = df.copy()
    # mapear columnas a esquema com√∫n
    mapped = {}
    for target, cands in SYNONYMS.items():
        col = pick_column(df, cands)
        mapped[target] = col

    out = pd.DataFrame()
    out["brand"] = brand

    # copiar columnas si existen
    out["post_id"]      = df[mapped["post_id"]] if mapped["post_id"] else np.nan
    out["caption"]      = df[mapped["caption"]] if mapped["caption"] else np.nan
    out["published_at"] = df[mapped["published_at"]] if mapped["published_at"] else np.nan
    out["likes"]        = df[mapped["likes"]] if mapped["likes"] else np.nan
    out["comments"]     = df[mapped["comments"]] if mapped["comments"] else np.nan
    out["shares"]       = df[mapped["shares"]] if mapped["shares"] else np.nan
    out["views"]        = df[mapped["views"]] if mapped["views"] else np.nan
    out["url"]          = df[mapped["url"]] if mapped["url"] else np.nan

    # limpiar tipos
    out["likes"]    = coerce_numeric(out["likes"])
    out["comments"] = coerce_numeric(out["comments"])
    out["shares"]   = coerce_numeric(out["shares"])
    out["views"]    = coerce_numeric(out["views"])

    out["published_at"] = parse_datetime(out["published_at"])

    # m√©tricas derivadas (solo donde tiene sentido)
    out["interactions"] = out[["likes","comments","shares"]].sum(axis=1, min_count=1)
    out["engagement_rate"] = out["interactions"] / out["views"].replace(0, np.nan)

    # sanity: evitar negativos raros
    for c in ["likes","comments","shares","views","interactions"]:
        out.loc[out[c] < 0, c] = np.nan

    return out, mapped

shein_df, shein_map = standardize(shein, "Shein")
temu_df,  temu_map  = standardize(temu,  "Temu")

df = pd.concat([shein_df, temu_df], ignore_index=True)

print("‚úÖ Tabla unificada df creada")
print("df shape:", df.shape)
display(df.head())
print("\nMapeo SHEIN:", shein_map)
print("Mapeo TEMU :", temu_map)



def fmt_int(x):
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return "NA"
    return f"{int(x):,}".replace(",", ".")

def fmt_float(x, nd=3):
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return "NA"
    return f"{x:.{nd}f}"

def fmt_pct(x, nd=1):
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return "NA"
    return f"{x*100:.{nd}f}%"

# 1) ¬øCu√°ntas publicaciones?
counts = df["brand"].value_counts()
n_shein = int(counts.get("Shein", 0))
n_temu  = int(counts.get("Temu", 0))

# 2) Totales y promedios (normalizados por post)
agg = df.groupby("brand").agg(
    posts=("brand","size"),
    total_likes=("likes","sum"),
    avg_likes=("likes","mean"),
    med_likes=("likes","median"),
    total_comments=("comments","sum"),
    avg_comments=("comments","mean"),
    total_shares=("shares","sum"),
    avg_shares=("shares","mean"),
    total_views=("views","sum"),
    avg_views=("views","mean"),
    views_coverage=("views", lambda s: float(s.notna().mean())),
    er_coverage=("engagement_rate", lambda s: float(s.notna().mean())),
    avg_er=("engagement_rate","mean"),
    med_er=("engagement_rate","median"),
)
display(agg)

# 3) Representatividad temporal (si published_at existe)
def time_window(sub):
    s = sub["published_at"]
    if s.notna().sum() < 5:
        return None
    return (s.min(), s.max())

tw_shein = time_window(df[df.brand=="Shein"])
tw_temu  = time_window(df[df.brand=="Temu"])

overlap_info = None
if tw_shein and tw_temu:
    start = max(tw_shein[0], tw_temu[0])
    end   = min(tw_shein[1], tw_temu[1])
    overlap_days = (end - start).days
    overlap_info = (start, end, overlap_days)

# 4) Tasa de publicaci√≥n (posts/d√≠a) si hay fechas
def posts_per_day(sub):
    tw = time_window(sub)
    if not tw:
        return None
    days = max(1, (tw[1] - tw[0]).days + 1)
    return float(len(sub)/days)

ppd_shein = posts_per_day(df[df.brand=="Shein"])
ppd_temu  = posts_per_day(df[df.brand=="Temu"])

# 5) Markdown autogenerado para pegar como celda de texto
shein_row = agg.loc["Shein"] if "Shein" in agg.index else None
temu_row  = agg.loc["Temu"]  if "Temu"  in agg.index else None


```

## 3. Limpieza de M√©tricas  
**Objetivo de negocio:** asegurar que las m√©tricas que usamos representen realidad y no errores de scraping o outliers raros.

### 3.1. Preguntas de negocio

1. **Datos faltantes**
   - ¬øQu√© porcentaje de posts tiene valores faltantes en m√©tricas clave (ej. `likes`, `comments`, `views`)?  
   - Desde el punto de vista de negocio, ¬øqu√© har√≠as con esos posts: los excluir√≠as, los imputar√≠as o los analizar√≠as aparte? ¬øPor qu√©?

2. **Outliers**
   - Identifica algunos posts con n√∫meros extremadamente altos de likes o views.  
   - ¬øCrees que son org√°nicos o podr√≠an ser campa√±as pagas / virales espec√≠ficos?  
   - ¬øC√≥mo cambia tu lectura si incluyes vs. excluyes estos outliers?

3. **Calidad de la m√©trica**
   - Si un post tiene 0 comments pero muchos likes, ¬øqu√© hip√≥tesis de negocio podr√≠as plantear sobre ese tipo de contenido o audiencia?  
   - ¬øQu√© te preocupa m√°s como inversor: muchos likes con poco comentario, o pocos likes pero una tasa de comentarios alta? Explica.


```python
import numpy as np
import pandas as pd

key_metrics = ["likes", "comments", "views", "shares", "interactions", "engagement_rate"]

# Asegurar que existan (por si el dataset no trae algunas)
key_metrics = [c for c in key_metrics if c in df.columns]

# % faltantes global
missing_global = (df[key_metrics].isna().mean() * 100).sort_values(ascending=False).round(2)
display(pd.DataFrame({"missing_%": missing_global}))

# % faltantes por marca
missing_by_brand = (df.groupby("brand")[key_metrics].apply(lambda g: g.isna().mean()*100)).round(2)
display(missing_by_brand)

# % de posts con al menos 1 m√©trica clave faltante (likes/comments/views)
base_keys = [c for c in ["likes","comments","views"] if c in df.columns]
df["has_missing_core"] = df[base_keys].isna().any(axis=1)
core_missing_rate = df["has_missing_core"].mean() * 100

core_missing_by_brand = (df.groupby("brand")["has_missing_core"].mean() * 100).round(2)

print(f"% posts con al menos 1 m√©trica core faltante ({base_keys}): {core_missing_rate:.2f}%")
display(core_missing_by_brand)


def iqr_bounds(s, k=1.5):
    s = s.dropna()
    if len(s) < 10:
        return (None, None)
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    low = q1 - k * iqr
    high = q3 + k * iqr
    return low, high

def mark_outliers(df, col, k=1.5):
    low, high = iqr_bounds(df[col], k=k)
    flag = pd.Series(False, index=df.index)
    if high is None:
        return flag, (low, high)
    flag = df[col] > high
    return flag, (low, high)

out_cols = [c for c in ["likes", "views"] if c in df.columns]

bounds = {}
for c in out_cols:
    flag, (low, high) = mark_outliers(df, c, k=1.5)
    df[f"outlier_{c}"] = flag
    bounds[c] = {"low": low, "high": high}

print("Bounds IQR (k=1.5):", bounds)

# Mostrar top 10 por likes y views (con columnas √∫tiles)
show_cols = [c for c in ["brand","post_id","published_at","likes","comments","shares","views","engagement_rate","url","caption"] if c in df.columns]

if "likes" in df.columns:
    display(df.sort_values("likes", ascending=False)[show_cols].head(10))

if "views" in df.columns:
    display(df.sort_values("views", ascending=False)[show_cols].head(10))

# Conteos de outliers por marca
for c in out_cols:
    out_rate = (df.groupby("brand")[f"outlier_{c}"].mean() * 100).round(2)
    print(f"\n% outliers en {c} por marca:")
    display(out_rate)

def summarize(df_in, label):
    g = df_in.groupby("brand").agg(
        posts=("brand","size"),
        avg_likes=("likes","mean"),
        med_likes=("likes","median"),
        avg_views=("views","mean") if "views" in df_in.columns else ("brand","size"),
        med_views=("views","median") if "views" in df_in.columns else ("brand","size"),
        avg_er=("engagement_rate","mean") if "engagement_rate" in df_in.columns else ("brand","size"),
        med_er=("engagement_rate","median") if "engagement_rate" in df_in.columns else ("brand","size"),
    )
    g["scenario"] = label
    return g.reset_index()

# Con outliers (baseline)
base = summarize(df, "incluye_outliers")

# Excluyendo outliers de likes y views (si existen)
mask = pd.Series(True, index=df.index)
if "outlier_likes" in df.columns:
    mask &= ~df["outlier_likes"]
if "outlier_views" in df.columns:
    mask &= ~df["outlier_views"]

df_no_out = df[mask].copy()
no_out = summarize(df_no_out, "excluye_outliers")

compare = pd.concat([base, no_out], ignore_index=True)
display(compare)

# Diferencia porcentual (promedios) para ver impacto
def pct_change(a, b):
    # (b-a)/a
    if a is None or b is None or (isinstance(a,float) and np.isnan(a)) or (isinstance(b,float) and np.isnan(b)) or a == 0:
        return np.nan
    return (b - a) / a

impact_rows = []
for brand in compare["brand"].unique():
    a = compare[(compare.brand==brand) & (compare.scenario=="incluye_outliers")].iloc[0]
    b = compare[(compare.brand==brand) & (compare.scenario=="excluye_outliers")].iloc[0]
    impact_rows.append({
        "brand": brand,
        "posts_before": int(a["posts"]),
        "posts_after": int(b["posts"]),
        "avg_likes_change_%": 100*pct_change(a["avg_likes"], b["avg_likes"]),
        "avg_views_change_%": 100*pct_change(a.get("avg_views", np.nan), b.get("avg_views", np.nan)),
        "avg_er_change_%": 100*pct_change(a.get("avg_er", np.nan), b.get("avg_er", np.nan)),
    })

impact = pd.DataFrame(impact_rows).round(2)
display(impact)


# Definir "muchos likes" como percentil 90 (puedes subir a 95)
p90_likes = df["likes"].quantile(0.90) if "likes" in df.columns else None

if p90_likes is not None and "comments" in df.columns:
    weird = df[(df["comments"].fillna(0) == 0) & (df["likes"].fillna(0) >= p90_likes)].copy()
    cols = [c for c in ["brand","post_id","likes","comments","views","engagement_rate","caption","url"] if c in df.columns]
    print("p90 likes =", p90_likes)
    print("Casos 0 comments y likes altos:", weird.shape[0])
    display(weird.sort_values("likes", ascending=False)[cols].head(15))
else:
    print("No existe likes o comments para este an√°lisis.")

# Comparar: muchos likes con poco comentario VS pocos likes con alta tasa de comentarios
if "likes" in df.columns and "comments" in df.columns:
    df["comment_rate"] = df["comments"] / df["likes"].replace(0, np.nan)

    high_likes_low_comments = df[(df["likes"] >= df["likes"].quantile(0.90)) & (df["comments"] <= df["comments"].quantile(0.25))]
    low_likes_high_commentrate = df[(df["likes"] <= df["likes"].quantile(0.25)) & (df["comment_rate"] >= df["comment_rate"].quantile(0.90))]

    print("Segmento A: muchos likes, pocos comments:", high_likes_low_comments.shape[0])
    print("Segmento B: pocos likes, alta tasa comments/likes:", low_likes_high_commentrate.shape[0])



def pct(x, nd=2):
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return "NA"
    return f"{x:.{nd}f}%"

def fnum(x, nd=3):
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return "NA"
    return f"{x:.{nd}f}"

core_missing = float(df["has_missing_core"].mean()*100) if "has_missing_core" in df.columns else np.nan

# Outlier rates
ol_likes = (df.groupby("brand")["outlier_likes"].mean()*100).to_dict() if "outlier_likes" in df.columns else {}
ol_views = (df.groupby("brand")["outlier_views"].mean()*100).to_dict() if "outlier_views" in df.columns else {}

# Impact table (si existe)
impact_dict = {}
if "impact" in globals():
    for _, r in impact.iterrows():
        impact_dict[r["brand"]] = r.to_dict()

md = f"""
## 3. Limpieza de M√©tricas

### 1) Datos faltantes
- **% de posts con al menos una m√©trica core faltante ({[c for c in ["likes","comments","views"] if c in df.columns]}):** **{core_missing:.2f}%**.
- **Qu√© har√≠a como negocio (decisi√≥n):**
  - Para an√°lisis comparativos entre marcas, **no imputar√≠a** likes/comments/views con 0 porque puede mezclar ‚Äúfaltante‚Äù con ‚Äúcero real‚Äù.
  - Har√≠a **dos cortes**:
    1) **Dataset completo** para volumen y cobertura.
    2) **Subset comparable** (solo posts con m√©tricas core presentes) para comparar rendimiento (promedios/medianas/ER).
  - Los posts con faltantes los analizar√≠a **aparte** como ‚Äúcalidad de extracci√≥n‚Äù (posible problema de scraping o API).

### 2) Outliers
- Identificamos outliers con regla IQR (k=1.5) en likes/views.
- **% outliers en likes:** Shein = {pct(ol_likes.get("Shein"))} | Temu = {pct(ol_likes.get("Temu"))}
- **% outliers en views:** Shein = {pct(ol_views.get("Shein"))} | Temu = {pct(ol_views.get("Temu"))}

**¬øOrg√°nico vs campa√±a? (hip√≥tesis de negocio)**
- Si un post tiene views/likes extremadamente altos, puede ser:
  - **viral org√°nico**, influenciado por tendencia/formato,
  - **campa√±a paga** (boost),
  - **colaboraci√≥n con creator** (influencer),
  - o **error de captura** (por ejemplo, duplicados o parsing).

**¬øC√≥mo cambia la lectura si excluyes outliers?**
- La m√©trica m√°s afectada suele ser el **promedio**; la **mediana** es m√°s robusta.
- Si al excluir outliers los promedios bajan mucho, significa que el ‚Äú√©xito‚Äù estaba concentrado en pocos virales.

### 3) Calidad de la m√©trica
**Caso: 0 comments pero muchos likes**
Hip√≥tesis:
- Contenido ‚Äúde consumo r√°pido‚Äù (hook visual) que incentiva like pero no conversaci√≥n.
- Audiencia pasiva (scroll) o CTA sin pregunta.
- Moderaci√≥n/limitaci√≥n de comentarios o contenido no controversial.

**Como inversor, ¬øqu√© preocupa m√°s?**
- **Muchos likes con poco comment**: puede ser awareness pero **baja profundidad** (menos se√±al de intenci√≥n/comunidad).
- **Pocos likes pero alta tasa de comments**: menor alcance, pero **alta calidad** (engagement profundo, posibles se√±ales de conversi√≥n/fidelidad).
- En estrategia: si quiero **ventas**, valoro m√°s profundidad (comentarios) + ER; si quiero **visibilidad**, likes/views pesan m√°s.
"""

print(md)
```

```python
# =========================
# 5) ENRIQUECIMIENTO CON IA
# (content_type + sentiment) + an√°lisis del punto 5
# =========================

# --- 0) Instalaci√≥n + API KEY segura (NO hardcode) ---
!pip -q install --upgrade openai

import os, re, json, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from getpass import getpass
from openai import OpenAI

if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass("Pega tu OPENAI_API_KEY (no se ver√°): ")

client = OpenAI()

# --- 1) Asegurar df (si no existe, intenta construirlo desde shein.csv y temu.csv) ---
def _norm(col): return re.sub(r"[^a-z0-9]+", "", str(col).strip().lower())

SYNONYMS = {
    "post_id": ["post_id","id","videoid","video_id","awemeid","aweme_id","itemid","item_id","uri"],
    "caption": ["caption","desc","description","text","content","title","videodesc"],
    "published_at": ["published_at","publish_time","created_at","create_time","date","datetime","timestamp","time"],
    "likes": ["likes","likescount","likecount","diggcount"],
    "comments": ["comments","commentscount","commentcount"],
    "shares": ["shares","sharescount","sharecount"],
    "views": ["views","viewscount","viewcount","playcount","impressions"],
    "url": ["url","shareurl","link","videourl","posturl"],
}

def pick_column(df, candidates):
    norm_cols = {_norm(c): c for c in df.columns}
    for cand in candidates:
        key = _norm(cand)
        if key in norm_cols:
            return norm_cols[key]
    for c in df.columns:
        if _norm(c) in {_norm(x) for x in candidates}:
            return c
    return None

def coerce_numeric(series):
    if series is None: return series
    s = series.copy()
    if s.dtype == "object":
        s = (s.astype(str)
             .str.replace(",", "", regex=False)
             .str.replace(" ", "", regex=False)
             .str.replace("\u200b", "", regex=False))
    return pd.to_numeric(s, errors="coerce")

def parse_datetime(series):
    if series is None: return series
    s = series.copy()
    num = pd.to_numeric(s, errors="coerce")
    if num.notna().mean() > 0.5:
        v = num.dropna()
        if len(v) > 0:
            med = float(v.median())
            unit = "ms" if med > 1e12 else "s"
            return pd.to_datetime(num, unit=unit, errors="coerce", utc=True).dt.tz_convert(None)
    return pd.to_datetime(s, errors="coerce", utc=True).dt.tz_convert(None)

def standardize(raw, brand):
    raw = raw.copy()
    mapped = {k: pick_column(raw, v) for k, v in SYNONYMS.items()}
    out = pd.DataFrame({"brand": brand})
    out["post_id"]      = raw[mapped["post_id"]] if mapped["post_id"] else np.nan
    out["caption"]      = raw[mapped["caption"]] if mapped["caption"] else np.nan
    out["published_at"] = raw[mapped["published_at"]] if mapped["published_at"] else np.nan
    out["likes"]        = raw[mapped["likes"]] if mapped["likes"] else np.nan
    out["comments"]     = raw[mapped["comments"]] if mapped["comments"] else np.nan
    out["shares"]       = raw[mapped["shares"]] if mapped["shares"] else np.nan
    out["views"]        = raw[mapped["views"]] if mapped["views"] else np.nan
    out["url"]          = raw[mapped["url"]] if mapped["url"] else np.nan

    for c in ["likes","comments","shares","views"]:
        out[c] = coerce_numeric(out[c])

    out["published_at"] = parse_datetime(out["published_at"])
    out["interactions"] = out[["likes","comments","shares"]].sum(axis=1, min_count=1)
    out.loc[out["interactions"] < 0, "interactions"] = np.nan
    for c in ["likes","comments","shares","views"]:
        out.loc[out[c] < 0, c] = np.nan
    return out

if "df" not in globals():
    if os.path.exists("shein.csv") and os.path.exists("temu.csv"):
        shein_raw = pd.read_csv("shein.csv")
        temu_raw  = pd.read_csv("temu.csv")
        df = pd.concat([standardize(shein_raw, "Shein"), standardize(temu_raw, "Temu")], ignore_index=True)
    else:
        raise NameError("No existe df en memoria y no encontr√© shein.csv/temu.csv en el entorno.")

# --- 2) engagement_rate (seg√∫n enunciado) ---
if "engagement_rate" not in df.columns:
    if all(c in df.columns for c in ["likes","comments","views"]):
        df["engagement_rate"] = (df["likes"].fillna(0) + df["comments"].fillna(0)) / df["views"].replace(0, np.nan)
    else:
        df["engagement_rate"] = np.nan

# --- 3) LLM labeling (content_type + sentiment) en una sola llamada por batch (con cache) ---
CATEGORIES = [
    "Haul","Meme","Descuento","Sostenibilidad","UGC",
    "Producto individual","Tutorial/HowTo","Sorteo/Giveaway",
    "Otros","Sin texto"
]
SENTIMENTS = ["positivo","neutral","negativo"]
ALLOWED_CT = set(CATEGORIES)
ALLOWED_SENT = set(SENTIMENTS)

def clean_caption(x, max_len=450):
    if pd.isna(x): return ""
    s = str(x)
    s = re.sub(r"http\S+|www\.\S+", " ", s)
    s = " ".join(s.split()).strip()
    return s[:max_len]

def extract_json(text):
    if not text: return None
    t = text.strip()
    # intenta recortar a [...] si existe
    if "[" in t and "]" in t:
        t2 = t[t.find("["):t.rfind("]")+1]
        try: return json.loads(t2)
        except Exception: pass
    # intenta recortar a {...} si existe
    if "{" in t and "}" in t:
        t2 = t[t.find("{"):t.rfind("}")+1]
        try: return json.loads(t2)
        except Exception: pass
    try:
        return json.loads(t)
    except Exception:
        return None

def label_batch(captions, model="gpt-4o-mini"):
    items = [{"i": i, "caption": clean_caption(c)} for i, c in enumerate(captions)]
    prompt = f"""
Etiquetas permitidas:
- content_type: {CATEGORIES}
- sentiment: {SENTIMENTS}

Tarea:
Para cada item, asigna:
1) content_type (UNA sola etiqueta exacta de la lista)
2) sentiment (positivo/neutral/negativo)

Reglas:
- Si caption est√° vac√≠o -> content_type="Sin texto" y sentiment="neutral"
- Responde SOLO JSON v√°lido (sin texto extra).
- Formato exacto: {{"items":[{{"i":0,"content_type":"...","sentiment":"..."}}, ...]}}
"""
    resp = client.responses.create(
        model=model,
        input=prompt + "\n\nCAPTIONS:\n" + json.dumps({"items": items}, ensure_ascii=False)
    )
    raw = getattr(resp, "output_text", "") or ""
    data = extract_json(raw)

    # fallback si no vino bien
    if not isinstance(data, dict) or "items" not in data:
        # segundo intento (m√°s estricto)
        resp2 = client.responses.create(
            model=model,
            input="Responde SOLO JSON v√°lido, sin texto extra.\n" + prompt + "\n\nCAPTIONS:\n" + json.dumps({"items": items}, ensure_ascii=False)
        )
        raw2 = getattr(resp2, "output_text", "") or ""
        data = extract_json(raw2)

    out_ct = ["Otros"] * len(captions)
    out_sent = ["neutral"] * len(captions)

    if isinstance(data, dict) and isinstance(data.get("items"), list):
        for row in data["items"]:
            try:
                i = int(row.get("i"))
                ct = str(row.get("content_type", "Otros")).strip()
                se = str(row.get("sentiment", "neutral")).strip().lower()
                out_ct[i] = ct if ct in ALLOWED_CT else "Otros"
                out_sent[i] = se if se in ALLOWED_SENT else "neutral"
            except Exception:
                continue
    return out_ct, out_sent

# Control de costo/tiempo: por defecto etiqueta hasta N filas
MAX_ROWS_FOR_LLM = None   # <-- si quieres limitar (ej. 400), pon 400
BATCH = 25
SLEEP = 0.15

work = df.copy()
if MAX_ROWS_FOR_LLM is not None and len(work) > MAX_ROWS_FOR_LLM:
    work = work.sample(MAX_ROWS_FOR_LLM, random_state=42).copy()

if "caption" not in work.columns:
    work["caption"] = ""

caps = work["caption"].fillna("").astype(str).map(clean_caption).tolist()

cache = {}  # caption -> (content_type, sentiment)
content_type_llm = [None] * len(caps)
sentiment_llm = [None] * len(caps)

# agrupar duplicados para ahorrar
unique_map = {}
for idx, cap in enumerate(caps):
    unique_map.setdefault(cap, []).append(idx)

uniq_caps = list(unique_map.keys())

for start in range(0, len(uniq_caps), BATCH):
    batch_caps = uniq_caps[start:start+BATCH]

    # resolver por cache primero
    to_call = []
    call_pos = []
    batch_ct = [None] * len(batch_caps)
    batch_se = [None] * len(batch_caps)

    for j, cap in enumerate(batch_caps):
        if cap in cache:
            batch_ct[j], batch_se[j] = cache[cap]
        elif cap.strip() == "":
            batch_ct[j], batch_se[j] = "Sin texto", "neutral"
            cache[cap] = (batch_ct[j], batch_se[j])
        else:
            to_call.append(cap)
            call_pos.append(j)

    if to_call:
        pred_ct, pred_se = label_batch(to_call)
        for k, j in enumerate(call_pos):
            batch_ct[j] = pred_ct[k]
            batch_se[j] = pred_se[k]
            cache[batch_caps[j]] = (batch_ct[j], batch_se[j])
        time.sleep(SLEEP)

    # asignar a todas las filas con ese caption
    for cap, ct, se in zip(batch_caps, batch_ct, batch_se):
        for idx in unique_map[cap]:
            content_type_llm[idx] = ct
            sentiment_llm[idx] = se

work["content_type_llm"] = content_type_llm
work["sentiment_llm"] = sentiment_llm

# copiar de vuelta a df (merge por √≠ndice original si usaste sample)
df = df.copy()
df.loc[work.index, "content_type_llm"] = work["content_type_llm"]
df.loc[work.index, "sentiment_llm"] = work["sentiment_llm"]

# Si no existe content_type ‚Äúoficial‚Äù, √∫salo desde LLM
if "content_type" not in df.columns or df["content_type"].isna().all():
    df["content_type"] = df["content_type_llm"]

# --- 4) An√°lisis del punto 5 (mix, performance, gap, recomendaci√≥n, greenwashing) ---
use_ct = "content_type"  # o "content_type_llm"
use_sent = "sentiment_llm" if "sentiment_llm" in df.columns else None

# Mix content_type
mix = df.dropna(subset=["brand", use_ct]).groupby(["brand", use_ct]).size().reset_index(name="posts")
mix["share"] = mix.groupby("brand")["posts"].apply(lambda s: s / s.sum())

mix_pivot = (mix.pivot_table(index=use_ct, columns="brand", values="share", fill_value=0) * 100).round(2)
print("\n=== MIX (%) por marca ===")
display(mix_pivot.sort_values(by=mix_pivot.columns.tolist(), ascending=False))

# Performance por content_type
perf = df.dropna(subset=["brand", use_ct]).groupby(["brand", use_ct]).agg(
    posts=("brand","size"),
    er_mean=("engagement_rate","mean"),
    er_median=("engagement_rate","median"),
    er_p75=("engagement_rate", lambda s: s.quantile(0.75)),
).reset_index()

perf = perf.merge(mix[["brand", use_ct, "share"]].rename(columns={"share":"mix_share"}), on=["brand", use_ct], how="left")
print("\n=== TOP content_type por ER (mediana) ===")
display(perf.sort_values(["brand","er_median"], ascending=[True, False]).groupby("brand").head(10))

# Gap estrategia vs performance (infra/sobre-invertido)
def rank_gap(sub):
    sub = sub.copy()
    sub["rank_er"] = sub["er_median"].rank(ascending=False, method="min")
    sub["rank_share"] = sub["mix_share"].rank(ascending=False, method="min")
    sub["gap"] = sub["rank_share"] - sub["rank_er"]  # + => rinde mejor de lo que inviertes
    return sub.sort_values("gap", ascending=False)

gaps = perf.groupby("brand", group_keys=False).apply(rank_gap)
print("\n=== GAP (infra-invertidos arriba) ===")
display(gaps[["brand",use_ct,"posts","mix_share","er_median","gap"]].groupby("brand").head(8))

# Recomendaci√≥n de mix futuro (marca con peor ER mediana)
brand_er = df.groupby("brand")["engagement_rate"].median().sort_values()
worst_brand = brand_er.index[0]
best_brand  = brand_er.index[-1]
print("\nPeor marca por ER (mediana):", worst_brand, "| Mejor:", best_brand)

wb = perf[perf.brand==worst_brand].dropna(subset=["er_median","mix_share"]).sort_values("er_median", ascending=False).copy()
top = wb.head(min(4, len(wb)))
bottom = wb.tail(min(3, len(wb)))

SHIFT = 0.20  # mover 20% del mix (ajustable)
proposal = wb[[use_ct,"mix_share","er_median"]].copy()
proposal["new_share"] = proposal["mix_share"]

bottom_total = bottom["mix_share"].sum()
top_weight = top["er_median"].sum()

for _, r in bottom.iterrows():
    cut = (r["mix_share"]/bottom_total) * SHIFT if bottom_total > 0 else 0
    proposal.loc[proposal[use_ct]==r[use_ct], "new_share"] = np.maximum(
        0, proposal.loc[proposal[use_ct]==r[use_ct], "new_share"] - cut
    )

for _, r in top.iterrows():
    add = (r["er_median"]/top_weight) * SHIFT if top_weight > 0 else 0
    proposal.loc[proposal[use_ct]==r[use_ct], "new_share"] += add

proposal["new_share"] = proposal["new_share"] / proposal["new_share"].sum()
proposal["mix_%"] = (proposal["mix_share"]*100).round(2)
proposal["new_mix_%"] = (proposal["new_share"]*100).round(2)

print("\n=== Propuesta de mix (%) para", worst_brand, "===")
display(proposal.sort_values("new_share", ascending=False)[[use_ct,"mix_%","new_mix_%","er_median"]])

# Greenwashing (opcional)
GREEN_KWS = ["eco","green","sustain","sostenibl","recicl","reutil","biodegrad","carbon","huella","clima","ambiente","medio ambiente","ethical","√©tica","etica"]
def is_green(x):
    s = clean_caption(x).lower()
    return any(k in s for k in GREEN_KWS)

df["is_green"] = df["caption"].apply(is_green)

green_share = (df.groupby("brand")["is_green"].mean() * 100).round(2)
green_perf = df[df["is_green"]==True].groupby("brand")["engagement_rate"].agg(["mean","median","count"])
overall_perf = df.groupby("brand")["engagement_rate"].agg(["mean","median","count"])

print("\n=== % contenido sostenible (keywords) ===")
display(green_share.rename("green_%"))

print("\n=== Engagement contenido sostenible vs total ===")
comp = overall_perf[["mean","median","count"]].rename(columns={"mean":"overall_mean","median":"overall_median","count":"overall_posts"}).join(
    green_perf[["mean","median","count"]].rename(columns={"mean":"green_mean","median":"green_median","count":"green_posts"}),
    how="left"
)
display(comp)

# Sentiment mix/performance y cruce con content_type (complemento)
if use_sent and use_sent in df.columns:
    mix_s = df.dropna(subset=["brand",use_sent]).groupby(["brand",use_sent]).size().reset_index(name="posts")
    mix_s["share"] = mix_s.groupby("brand")["posts"].apply(lambda s: s/s.sum())
    print("\n=== MIX sentiment (%) ===")
    display((mix_s.pivot_table(index=use_sent, columns="brand", values="share", fill_value=0)*100).round(2))

    perf_s = df.dropna(subset=["brand",use_sent]).groupby(["brand",use_sent]).agg(
        posts=("brand","size"),
        er_median=("engagement_rate","median"),
        er_mean=("engagement_rate","mean"),
    ).reset_index().sort_values(["brand","er_median"], ascending=[True, False])
    print("\n=== Performance por sentiment ===")
    display(perf_s)

    combo = df.dropna(subset=["brand",use_ct,use_sent]).groupby(["brand",use_ct,use_sent]).agg(
        posts=("brand","size"),
        er_median=("engagement_rate","median"),
        er_mean=("engagement_rate","mean"),
    ).reset_index()
    combo = combo[combo["posts"] >= 5].sort_values(["brand","er_median"], ascending=[True, False])
    print("\n=== TOP combinaciones content_type x sentiment (posts>=5) ===")
    display(combo.groupby("brand").head(15))

# --- 5) Markdown final (respuestas de negocio del punto 5) ---
def pct(x, nd=1):
    if x is None or (isinstance(x, float) and np.isnan(x)): return "NA"
    return f"{x*100:.{nd}f}%"

def f(x, nd=4):
    if x is None or (isinstance(x, float) and np.isnan(x)): return "NA"
    return f"{x:.{nd}f}"

brands = df["brand"].dropna().unique().tolist()
mix_top = mix.sort_values(["brand","share"], ascending=[True, False]).groupby("brand").head(5)
best_ct = perf.sort_values(["brand","er_median"], ascending=[True, False]).groupby("brand").head(5)

infra = gaps.sort_values(["brand","gap"], ascending=[True, False]).groupby("brand").head(3)
over  = gaps.sort_values(["brand","gap"], ascending=[True, True]).groupby("brand").head(3)

md = "## 5. Enriquecimiento con IA (Categorizaci√≥n Sem√°ntica)\n\n"
md += "**Objetivo:** transformar captions a insight estrat√©gico sobre tipos de contenido.\n\n"

md += "### 1) Mix de contenido\n"
for b in brands:
    rows = mix_top[mix_top.brand==b][[use_ct,"share"]].values.tolist()
    md += f"- **{b}:** " + ", ".join([f"{ct} ({pct(sh)})" for ct, sh in rows]) + "\n"

md += "\n**Posicionamiento (interpretaci√≥n):**\n"
md += "- M√°s **Descuento** = performance/precio.\n"
md += "- M√°s **UGC/Haul** = prueba social y confianza.\n"
md += "- M√°s **Meme/Trend** = alcance/afinidad cultural.\n"
md += "- **Sostenibilidad** = narrativa de marca (validar si rinde vs promedio).\n\n"

md += "### 2) Contenido que mejor funciona (engagement)\n"
for b in brands:
    rows = best_ct[best_ct.brand==b][[use_ct,"er_median"]].values.tolist()
    md += f"- **{b}:** " + ", ".join([f"{ct} (med={f(er)})" for ct, er in rows]) + "\n"

md += "\n**¬øEst√°n invirtiendo donde rinde? (gap estrategia vs performance):**\n"
for b in brands:
    a = infra[infra.brand==b][[use_ct,"gap"]].values.tolist()
    d = over[over.brand==b][[use_ct,"gap"]].values.tolist()
    md += f"- **{b} infra-invertidos:** " + ", ".join([f"{ct}" for ct, _ in a]) + "\n"
    md += f"  **{b} sobre-invertidos:** " + ", ".join([f"{ct}" for ct, _ in d]) + "\n"

md += "\n### 3) Ajuste estrategia-audiencia\n"
md += "- **Memes/Trends** suelen atraer audiencia de descubrimiento (top-of-funnel).\n"
md += "- **Hauls/UGC** suelen atraer audiencia con intenci√≥n (mid-funnel) por validaci√≥n y prueba social.\n"
md += "- Lo m√°s sostenible a largo plazo suele ser: **consistencia (UGC/HowTo) + alcance (Trends)**.\n\n"

md += "### 4) Recomendaci√≥n de mix futuro\n"
md += f"- **Marca con peor ER mediano:** **{worst_brand}**.\n"
md += "- Propuesta: mover parte del mix desde formatos de bajo ER hacia los top performers y validar A/B.\n"
md += "- **Mix propuesto (top 10):**\n"
for _, r in proposal.sort_values("new_share", ascending=False).head(10).iterrows():
    md += f"  - {r[use_ct]}: {r['mix_%']}% ‚Üí {r['new_mix_%']}%\n"

md += "\n**A/B sugeridos:**\n"
md += "- M√°s **UGC** vs m√°s **Descuento**.\n"
md += "- **Haul corto** vs **haul largo**.\n"
md += "- Trend con producto integrado vs trend sin producto.\n\n"

md += "### 5) Greenwashing (opcional)\n"
for b in brands:
    g = float(green_share.get(b, np.nan))/100 if b in green_share.index else np.nan
    md += f"- **{b}:** contenido sostenible aprox = **{pct(g)}**.\n"
md += "- Si el engagement del contenido sostenible est√° por debajo del promedio, puede ser se√±al de baja credibilidad o mala ejecuci√≥n del mensaje.\n"

print("\n" + md)
```

## 6. An√°lisis Temporal (Heatmap)  
**Objetivo:** encontrar ‚Äúventanas de oro‚Äù de publicaci√≥n combinando d√≠a y hora.

Sup√≥n que ya creaste variables como `day_of_week` y `hour` y un heatmap de `engagement_rate` (o `likes`) por d√≠a y hora.

### 6.1. Preguntas de negocio

1. **Hora de oro por marca**
   - ¬øEn qu√© d√≠as y horarios Shein obtiene mejor resultado promedio?  
   - ¬øY Temu? ¬øCoinciden esos ‚Äúsweet spots‚Äù o est√°n jugando a momentos distintos del d√≠a?

2. **Volumen vs performance en el tiempo**
   - ¬øHay franjas horarias donde la marca publica mucho pero el engagement no es particularmente alto?  
   - Si tuvieras que recortar presupuesto/horas, ¬øqu√© franjas eliminar√≠as primero?

3. **Oportunidades desatendidas**
   - ¬øVes franjas donde se publica poco pero el engagement promedio es alto?  
   - Si fueras el CMO, ¬øprobar√≠as aumentar la inversi√≥n en esas ventanas aunque hoy casi no haya posts?

4. **Decisi√≥n de campa√±a clave**
   - Si el fondo decide financiar una gran campa√±a puntual (por ejemplo, lanzamiento de colecci√≥n o gran descuento):  
     - ¬øEn qu√© d√≠a y hora la lanzar√≠as para maximizar impacto org√°nico?  
     - Justifica usando el heatmap (no solo tu intuici√≥n).


```python
# =========================
# 6) AN√ÅLISIS TEMPORAL (HEATMAP)
# Ventanas de oro: d√≠a + hora vs engagement_rate (o likes)
# =========================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- 0) Validaciones b√°sicas ----------
if "published_at" not in df.columns:
    raise ValueError("No existe 'published_at' en df. Necesitas la fecha/hora de publicaci√≥n.")
if "brand" not in df.columns:
    raise ValueError("No existe 'brand' en df.")

# Asegurar datetime
df = df.copy()
df["published_at"] = pd.to_datetime(df["published_at"], errors="coerce")

# M√©trica objetivo (usa engagement_rate si existe, si no usa likes)
metric = "engagement_rate" if "engagement_rate" in df.columns and df["engagement_rate"].notna().any() else "likes"
if metric not in df.columns:
    raise ValueError("No existe 'engagement_rate' ni 'likes' para construir el heatmap.")

# ---------- 1) Features: day_of_week y hour ----------
# day_of_week: 0=lunes ... 6=domingo (m√°s negocio-friendly)
df["day_of_week"] = df["published_at"].dt.dayofweek
df["hour"] = df["published_at"].dt.hour

dow_names = ["Lunes","Martes","Mi√©rcoles","Jueves","Viernes","S√°bado","Domingo"]
df["day_name"] = df["day_of_week"].map(lambda x: dow_names[int(x)] if pd.notna(x) else np.nan)

# Filtrar filas v√°lidas
temp = df.dropna(subset=["brand","day_of_week","hour",metric]).copy()
temp["day_of_week"] = temp["day_of_week"].astype(int)
temp["hour"] = temp["hour"].astype(int)

# ---------- 2) Heatmap por marca (promedio) + conteo de posts ----------
def heatmap_table(data, metric_col):
    # promedio y conteo para el mismo grid
    pivot_mean = data.pivot_table(index="day_of_week", columns="hour", values=metric_col, aggfunc="mean")
    pivot_cnt  = data.pivot_table(index="day_of_week", columns="hour", values=metric_col, aggfunc="size")
    # reindex para que salga Lunes..Domingo y horas 0..23
    pivot_mean = pivot_mean.reindex(range(7)).reindex(columns=range(24))
    pivot_cnt  = pivot_cnt.reindex(range(7)).reindex(columns=range(24))
    return pivot_mean, pivot_cnt

def plot_heatmap(pivot, title, vmin=None, vmax=None):
    plt.figure(figsize=(14,4.8))
    plt.imshow(pivot.values, aspect="auto", interpolation="nearest", vmin=vmin, vmax=vmax)
    plt.yticks(range(7), dow_names)
    plt.xticks(range(24), range(24))
    plt.xlabel("Hora del d√≠a")
    plt.ylabel("D√≠a de la semana")
    plt.title(title)
    plt.colorbar(label=metric)
    plt.show()

brands = temp["brand"].unique().tolist()

heat = {}
for b in brands:
    sub = temp[temp["brand"]==b]
    hm_mean, hm_cnt = heatmap_table(sub, metric)
    heat[b] = {"mean": hm_mean, "count": hm_cnt}

    # Escala por marca (independiente) para ver su patr√≥n interno
    plot_heatmap(hm_mean, f"Heatmap {metric} promedio - {b}")

    # Heatmap de volumen (conteo)
    plot_heatmap(hm_cnt.fillna(0), f"Heatmap volumen (#posts) - {b}", vmin=0)

# ---------- 3) Sweet spots: top celdas por marca (con m√≠nimo de posts) ----------
MIN_POSTS_CELL = 5  # ajusta (ej: 3 si tienes pocos datos)

def top_cells(hm_mean, hm_cnt, topk=10, min_posts=5):
    rows = []
    for d in range(7):
        for h in range(24):
            m = hm_mean.loc[d, h]
            c = hm_cnt.loc[d, h]
            if pd.notna(m) and pd.notna(c) and c >= min_posts:
                rows.append((d, h, float(c), float(m)))
    rows.sort(key=lambda x: x[3], reverse=True)
    return pd.DataFrame(rows[:topk], columns=["day_of_week","hour","posts","metric_mean"])

sweet_spots = {}
for b in brands:
    ss = top_cells(heat[b]["mean"], heat[b]["count"], topk=12, min_posts=MIN_POSTS_CELL)
    ss["day_name"] = ss["day_of_week"].map(lambda x: dow_names[int(x)])
    sweet_spots[b] = ss
    print(f"\n=== Top ventanas (sweet spots) - {b} | min_posts={MIN_POSTS_CELL} ===")
    display(ss[["day_name","hour","posts","metric_mean"]])

# ---------- 4) Volumen vs performance: detectar franjas con mucho post y bajo rendimiento ----------
# Idea: ‚Äúmucho volumen‚Äù = arriba del percentil 75 de conteo; ‚Äúbajo performance‚Äù = debajo de la mediana de la marca
def volume_vs_perf_flags(hm_mean, hm_cnt):
    cnt_vals = hm_cnt.values.flatten()
    mean_vals = hm_mean.values.flatten()
    cnt_vals = cnt_vals[~np.isnan(cnt_vals)]
    mean_vals = mean_vals[~np.isnan(mean_vals)]
    if len(cnt_vals) == 0 or len(mean_vals) == 0:
        return None
    cnt_hi = np.nanpercentile(cnt_vals, 75)
    mean_lo = np.nanpercentile(mean_vals, 50)  # mediana
    flags = []
    for d in range(7):
        for h in range(24):
            c = hm_cnt.loc[d,h]
            m = hm_mean.loc[d,h]
            if pd.notna(c) and pd.notna(m) and c >= cnt_hi and m <= mean_lo:
                flags.append((d,h,float(c),float(m)))
    out = pd.DataFrame(flags, columns=["day_of_week","hour","posts","metric_mean"])
    if len(out):
        out["day_name"] = out["day_of_week"].map(lambda x: dow_names[int(x)])
        out = out.sort_values(["posts","metric_mean"], ascending=[False, True])
    return out, cnt_hi, mean_lo

low_eff_windows = {}
for b in brands:
    res = volume_vs_perf_flags(heat[b]["mean"], heat[b]["count"])
    if res is None:
        continue
    out, cnt_hi, mean_lo = res
    low_eff_windows[b] = out
    print(f"\n=== Mucho volumen pero bajo {metric} - {b} ===")
    print(f"Umbral volumen (p75 posts/celda) = {cnt_hi:.1f} | Umbral bajo performance (mediana) = {mean_lo:.6f}")
    display(out.head(15) if out is not None else out)

# ---------- 5) Oportunidades desatendidas: poco volumen y alto performance ----------
# Pocas publicaciones = <= percentil 25 de conteo, alto performance = >= percentil 75 de performance
def underused_opportunities(hm_mean, hm_cnt):
    cnt_vals = hm_cnt.values.flatten()
    mean_vals = hm_mean.values.flatten()
    cnt_vals = cnt_vals[~np.isnan(cnt_vals)]
    mean_vals = mean_vals[~np.isnan(mean_vals)]
    if len(cnt_vals) == 0 or len(mean_vals) == 0:
        return None
    cnt_lo = np.nanpercentile(cnt_vals, 25)
    mean_hi = np.nanpercentile(mean_vals, 75)
    opp = []
    for d in range(7):
        for h in range(24):
            c = hm_cnt.loc[d,h]
            m = hm_mean.loc[d,h]
            if pd.notna(c) and pd.notna(m) and c <= cnt_lo and m >= mean_hi:
                opp.append((d,h,float(c),float(m)))
    out = pd.DataFrame(opp, columns=["day_of_week","hour","posts","metric_mean"])
    if len(out):
        out["day_name"] = out["day_of_week"].map(lambda x: dow_names[int(x)])
        out = out.sort_values(["metric_mean","posts"], ascending=[False, True])
    return out, cnt_lo, mean_hi

opps = {}
for b in brands:
    res = underused_opportunities(heat[b]["mean"], heat[b]["count"])
    if res is None:
        continue
    out, cnt_lo, mean_hi = res
    opps[b] = out
    print(f"\n=== Oportunidades desatendidas (poco volumen, alto {metric}) - {b} ===")
    print(f"Umbral poco volumen (p25 posts/celda) = {cnt_lo:.1f} | Umbral alto performance (p75) = {mean_hi:.6f}")
    display(out.head(15) if out is not None else out)

# ---------- 6) Recomendar ‚Äúd√≠a/hora de campa√±a‚Äù (por marca y global) ----------
# Elegimos la mejor celda por performance, con soporte m√≠nimo (posts>=MIN_POSTS_CELL).
def best_campaign_window(ss_df):
    if ss_df is None or len(ss_df)==0:
        return None
    r = ss_df.iloc[0]
    return {"day_name": r["day_name"], "hour": int(r["hour"]), "posts": int(r["posts"]), "metric_mean": float(r["metric_mean"])}

campaign = {}
for b in brands:
    campaign[b] = best_campaign_window(sweet_spots.get(b))

# ‚ÄúGlobal‚Äù: si quisieras una sola ventana (promedio de marcas) -> promedio de heatmaps
if len(brands) >= 2:
    mean_stack = np.nanmean(np.stack([heat[b]["mean"].values for b in brands], axis=0), axis=0)
    cnt_stack  = np.nansum(np.stack([heat[b]["count"].fillna(0).values for b in brands], axis=0), axis=0)
    hm_mean_global = pd.DataFrame(mean_stack, index=range(7), columns=range(24))
    hm_cnt_global  = pd.DataFrame(cnt_stack,  index=range(7), columns=range(24))
    ss_global = top_cells(hm_mean_global, hm_cnt_global, topk=10, min_posts=MIN_POSTS_CELL)
    ss_global["day_name"] = ss_global["day_of_week"].map(lambda x: dow_names[int(x)])
    campaign["Global"] = best_campaign_window(ss_global)
else:
    campaign["Global"] = None

print("\n=== Recomendaci√≥n de ventana de campa√±a ===")
display(pd.DataFrame(campaign).T)

# ---------- 7) Markdown autogenerado para responder 6.1 ----------
def pct(x, nd=2):
    if x is None or (isinstance(x,float) and np.isnan(x)): return "NA"
    return f"{x*100:.{nd}f}%"

def f(x, nd=6):
    if x is None or (isinstance(x,float) and np.isnan(x)): return "NA"
    return f"{x:.{nd}f}"

md = "## 6. An√°lisis Temporal (Heatmap)\n"
md += f"**Objetivo:** encontrar ventanas de oro combinando d√≠a y hora usando `{metric}`.\n\n"

md += "### 1) Hora de oro por marca\n"
for b in brands:
    rec = campaign.get(b)
    if rec:
        md += f"- **{b}:** mejor ventana observada = **{rec['day_name']} {rec['hour']}:00** (promedio {metric}={f(rec['metric_mean'])}, soporte={rec['posts']} posts en esa celda).\n"
    else:
        md += f"- **{b}:** no hay suficiente soporte para recomendar (revisa fechas o baja MIN_POSTS_CELL).\n"

md += "\n**¬øCoinciden los sweet spots?**\n"
if all(campaign.get(b) for b in brands):
    same = len({(campaign[b]['day_name'], campaign[b]['hour']) for b in brands}) == 1
    md += "- " + ("S√≠, coinciden en la misma ventana" if same else "No, est√°n jugando a momentos distintos (diferente d√≠a/hora).") + ".\n"
else:
    md += "- No se pudo evaluar coincidencia por falta de soporte en alguna marca.\n"

md += "\n### 2) Volumen vs performance en el tiempo\n"
for b in brands:
    out = low_eff_windows.get(b)
    if out is not None and len(out):
        sample = out.iloc[0]
        md += f"- **{b}:** hay franjas con **mucho volumen** pero **bajo {metric}**, por ejemplo **{sample['day_name']} {int(sample['hour'])}:00** (posts={int(sample['posts'])}, {metric}={f(sample['metric_mean'])}).\n"
        md += f"  Si recorto esfuerzo/presupuesto, empezar√≠a por estas franjas (alta producci√≥n, bajo retorno promedio).\n"
    else:
        md += f"- **{b}:** no aparecen franjas claras de ‚Äúmucho volumen y bajo rendimiento‚Äù (o hay pocos datos).\n"

md += "\n### 3) Oportunidades desatendidas\n"
for b in brands:
    out = opps.get(b)
    if out is not None and len(out):
        sample = out.iloc[0]
        md += f"- **{b}:** se ven ventanas con **poco volumen** pero **alto {metric}**, por ejemplo **{sample['day_name']} {int(sample['hour'])}:00** (posts={int(sample['posts'])}, {metric}={f(sample['metric_mean'])}).\n"
        md += "  Como CMO, **s√≠** probar√≠a aumentar inversi√≥n ah√≠ (test controlado) porque el promedio sugiere eficiencia.\n"
    else:
        md += f"- **{b}:** no se detectaron oportunidades claras (o el umbral de posts m√≠nimos es alto).\n"

md += "\n### 4) Decisi√≥n de campa√±a clave\n"
glob = campaign.get("Global")
if glob:
    md += f"- Si debo lanzar una campa√±a puntual para maximizar impacto org√°nico, usar√≠a **{glob['day_name']} {glob['hour']}:00**.\n"
    md += f"  Justificaci√≥n: en el heatmap global es la celda con mayor `{metric}` promedio con soporte suficiente (posts={glob['posts']}).\n"
else:
    md += "- Si tengo que elegir una sola ventana global, usar√≠a el mejor sweet spot por marca (arriba) o bajar√≠a el m√≠nimo de soporte para calcular un global.\n"

print("\n" + md)
```

```python
# ===============================
# 7) ENTREGABLE EJECUTIVO (FINAL)
# Conecta: content_type + engagement + temporal + (greenwashing opcional)
# Imprime respuestas listas para pegar en Markdown
# ===============================

import numpy as np
import pandas as pd
import re

# ---------- Helpers ----------
DOW_NAMES = ["Lunes","Martes","Mi√©rcoles","Jueves","Viernes","S√°bado","Domingo"]

def _ensure_dt(s):
    return pd.to_datetime(s, errors="coerce")

def _pick_col(*cands):
    for c in cands:
        if c in df.columns:
            return c
    return None

def _fmt(x, nd=4):
    if x is None or (isinstance(x,float) and np.isnan(x)):
        return "NA"
    return f"{x:.{nd}f}"

def _pct(x, nd=1):
    if x is None or (isinstance(x,float) and np.isnan(x)):
        return "NA"
    return f"{x*100:.{nd}f}%"

def _safe_div(a,b):
    if b is None or (isinstance(b,float) and np.isnan(b)) or b == 0:
        return np.nan
    return a/b

def _top_list(rows, k=3):
    rows = rows[:k]
    return ", ".join(rows) if rows else "NA"

def _clean_text(x, max_len=260):
    if pd.isna(x): return ""
    s = str(x)
    s = re.sub(r"http\S+|www\.\S+", " ", s)
    s = " ".join(s.split()).strip()
    return s[:max_len]

# ---------- Base df ----------
df2 = df.copy()

# Datetime + features temporales
if "published_at" in df2.columns:
    df2["published_at"] = _ensure_dt(df2["published_at"])
    df2["day_of_week"] = df2["published_at"].dt.dayofweek
    df2["hour"] = df2["published_at"].dt.hour
    df2["day_name"] = df2["day_of_week"].map(lambda x: DOW_NAMES[int(x)] if pd.notna(x) else np.nan)
else:
    df2["day_of_week"] = np.nan
    df2["hour"] = np.nan
    df2["day_name"] = np.nan

# M√©trica principal
metric = "engagement_rate" if ("engagement_rate" in df2.columns and df2["engagement_rate"].notna().any()) else None
if metric is None and all(c in df2.columns for c in ["likes","comments","views"]):
    df2["engagement_rate"] = (df2["likes"].fillna(0) + df2["comments"].fillna(0)) / df2["views"].replace(0, np.nan)
    metric = "engagement_rate"
if metric is None and "likes" in df2.columns:
    metric = "likes"
if metric is None:
    raise ValueError("No encuentro engagement_rate ni likes para hacer el entregable.")

# content_type (prioriza LLM si existe)
ct_col = _pick_col("content_type_llm", "content_type")
if ct_col is None:
    df2["content_type"] = "Otros"
    ct_col = "content_type"

# ---------- 7.1: Estrategia (mix + performance + temporal) ----------
# Mix por marca
mix = df2.dropna(subset=["brand", ct_col]).groupby(["brand", ct_col]).size().reset_index(name="posts")
mix["share"] = mix.groupby("brand")["posts"].apply(lambda s: s / s.sum())

mix_top = (mix.sort_values(["brand","share"], ascending=[True,False])
             .groupby("brand").head(4))
mix_dict = {b: mix_top[mix_top.brand==b][[ct_col,"share"]].values.tolist()
            for b in df2["brand"].dropna().unique()}

# Engagement por categor√≠a (mediana)
perf_ct = (df2.dropna(subset=["brand", ct_col, metric])
             .groupby(["brand", ct_col])[metric]
             .agg(er_median="median", er_mean="mean", posts="size")
             .reset_index()
             .sort_values(["brand","er_median"], ascending=[True,False]))

perf_top = perf_ct.groupby("brand").head(4)
perf_dict = {b: perf_top[perf_top.brand==b][[ct_col,"er_median"]].values.tolist()
             for b in df2["brand"].dropna().unique()}

# Patr√≥n temporal: mejores ventanas por marca (si hay day/hour)
MIN_POSTS_CELL = 5
best_window = {}

if df2["day_of_week"].notna().any() and df2["hour"].notna().any():
    tvalid = df2.dropna(subset=["brand","day_of_week","hour",metric]).copy()
    tvalid["day_of_week"] = tvalid["day_of_week"].astype(int)
    tvalid["hour"] = tvalid["hour"].astype(int)

    for b in tvalid["brand"].unique():
        sub = tvalid[tvalid["brand"]==b]
        hm_mean = sub.pivot_table(index="day_of_week", columns="hour", values=metric, aggfunc="mean").reindex(range(7)).reindex(columns=range(24))
        hm_cnt  = sub.pivot_table(index="day_of_week", columns="hour", values=metric, aggfunc="size").reindex(range(7)).reindex(columns=range(24))

        # top celdas con soporte
        rows = []
        for d in range(7):
            for h in range(24):
                m = hm_mean.loc[d,h]
                c = hm_cnt.loc[d,h]
                if pd.notna(m) and pd.notna(c) and c >= MIN_POSTS_CELL:
                    rows.append((d,h,int(c),float(m)))
        rows.sort(key=lambda x: x[3], reverse=True)
        if rows:
            d,h,c,m = rows[0]
            best_window[b] = {"day": DOW_NAMES[d], "hour": h, "posts": c, "value": m}
        else:
            best_window[b] = None
else:
    for b in df2["brand"].dropna().unique():
        best_window[b] = None

# ---------- 7.2: Eficiencia vs volumen ----------
brand_summary = (df2.dropna(subset=["brand"])
                   .groupby("brand")
                   .agg(posts=("brand","size"),
                        metric_median=(metric,"median"),
                        metric_mean=(metric,"mean"))
                   .reset_index()
                   .sort_values("metric_median", ascending=False))

winner_eff = brand_summary.iloc[0]["brand"] if len(brand_summary) else "NA"
loser_eff  = brand_summary.iloc[-1]["brand"] if len(brand_summary) else "NA"

# ---------- 7.3: Ventana de oro conjunta A/B ----------
# Elegir la celda (d√≠a,hora) donde ambas marcas est√©n fuertes (score por z-score + soporte m√≠nimo)
ab_window = None

if df2["day_of_week"].notna().any() and df2["hour"].notna().any():
    tvalid = df2.dropna(subset=["brand","day_of_week","hour",metric]).copy()
    tvalid["day_of_week"] = tvalid["day_of_week"].astype(int)
    tvalid["hour"] = tvalid["hour"].astype(int)
    brands = sorted(tvalid["brand"].unique())
    if len(brands) >= 2:
        # construir grids por marca
        grids = {}
        for b in brands:
            sub = tvalid[tvalid["brand"]==b]
            m = sub.pivot_table(index="day_of_week", columns="hour", values=metric, aggfunc="mean").reindex(range(7)).reindex(columns=range(24))
            c = sub.pivot_table(index="day_of_week", columns="hour", values=metric, aggfunc="size").reindex(range(7)).reindex(columns=range(24))
            grids[b] = (m, c)

        # z-score por marca para comparar patrones
        def zscore(mat):
            v = mat.values.flatten()
            v = v[~np.isnan(v)]
            if len(v) < 10:
                return mat * np.nan
            mu, sig = np.nanmean(v), np.nanstd(v)
            sig = sig if sig > 0 else 1.0
            return (mat - mu) / sig

        z = {b: zscore(grids[b][0]) for b in brands}

        best = None
        for d in range(7):
            for h in range(24):
                ok = True
                score = 0.0
                support = 0
                for b in brands:
                    m,c = grids[b]
                    if pd.isna(m.loc[d,h]) or pd.isna(c.loc[d,h]) or c.loc[d,h] < MIN_POSTS_CELL:
                        ok = False
                        break
                    score += float(z[b].loc[d,h])
                    support += int(c.loc[d,h])
                if ok:
                    score = score / len(brands)
                    # ligera recompensa por soporte total
                    score += 0.02 * np.log1p(support)
                    if best is None or score > best[0]:
                        best = (score, d, h, support)
        if best:
            _, d, h, support = best
            ab_window = {"day": DOW_NAMES[d], "hour": h, "support_total": support}

# fallback si no hay ab_window (usar la mejor de la marca m√°s eficiente)
if ab_window is None:
    if winner_eff in best_window and best_window[winner_eff] is not None:
        bw = best_window[winner_eff]
        ab_window = {"day": bw["day"], "hour": bw["hour"], "support_total": bw["posts"]}
    else:
        ab_window = {"day": "NA", "hour": "NA", "support_total": "NA"}

# ---------- 7.4: Greenwashing detector (opcional) ----------
GREEN_KWS = ["eco","green","sustain","sostenibl","recicl","reutil","biodegrad","carbon","huella","clima","ambiente","medio ambiente","ethical","√©tica","etica"]
if "caption" not in df2.columns:
    df2["caption"] = ""

def is_green(text):
    s = _clean_text(text).lower()
    return any(k in s for k in GREEN_KWS)

df2["is_green"] = df2["caption"].apply(is_green)

green_share = df2.groupby("brand")["is_green"].mean()
green_perf  = df2[df2["is_green"]==True].groupby("brand")[metric].median()
overall_med = df2.groupby("brand")[metric].median()

green_flag = {}
for b in df2["brand"].dropna().unique():
    gsh = float(green_share.get(b, np.nan))
    gp  = float(green_perf.get(b, np.nan)) if b in green_perf.index else np.nan
    om  = float(overall_med.get(b, np.nan))
    # heur√≠stica: ‚Äúmucho green‚Äù y ‚Äúpeor que promedio‚Äù
    flag = (pd.notna(gsh) and gsh >= 0.10) and (pd.notna(gp) and pd.notna(om) and gp < 0.90*om)
    green_flag[b] = {"green_share": gsh, "green_median": gp, "overall_median": om, "possible_greenwashing": bool(flag)}

# ---------- 7.5: Recomendaci√≥n ejecutiva (1 p√°rrafo) ----------
# escoger apuesta principal: m√°s eficiencia (mediana) y consistencia (p75 vs median aprox)
pick_brand = winner_eff

# ajustes concretos: (1) mix -> subir top performers, (2) calendario -> ventana A/B, (3) KPI -> mediana ER + coverage + % posts top quartile
def best_types_for_brand(b, k=3):
    sub = perf_ct[perf_ct.brand==b].head(k)
    return sub[ct_col].tolist()

def most_common_types_for_brand(b, k=3):
    sub = mix[mix.brand==b].sort_values("share", ascending=False).head(k)
    return sub[ct_col].tolist()

adjust_mix_up = best_types_for_brand(pick_brand, 3)
adjust_mix_down = most_common_types_for_brand(pick_brand, 3)

# ---------- Respuestas (Markdown) ----------
brands = sorted(df2["brand"].dropna().unique())

def line_strategy(b):
    mix_lines = mix_dict.get(b, [])
    perf_lines = perf_dict.get(b, [])
    mix_str = ", ".join([f"{ct} ({_pct(sh)})" for ct, sh in mix_lines]) if mix_lines else "NA"
    perf_str = ", ".join([f"{ct} (med={_fmt(er)})" for ct, er in perf_lines]) if perf_lines else "NA"
    bw = best_window.get(b)
    time_str = f"{bw['day']} {bw['hour']}:00" if bw else "NA"
    return mix_str, perf_str, time_str

# 7.1 texto (5‚Äì7 l√≠neas)
s_lines = []
for b in brands:
    mix_str, perf_str, time_str = line_strategy(b)
    s_lines.append(f"- **{b}**: mix dominante = {mix_str}; lo que mejor rinde (por {metric} mediano) = {perf_str}; ventana fuerte = {time_str}.")

# 7.2
# ‚Äúganando eficiencia‚Äù: winner_eff
winner_row = brand_summary[brand_summary.brand==winner_eff].iloc[0] if len(brand_summary) else None
loser_row  = brand_summary[brand_summary.brand==loser_eff].iloc[0] if len(brand_summary) else None

# 7.3
ab_str = f"{ab_window['day']} {ab_window['hour']}:00" if ab_window else "NA"

# 7.4 texto
gw_lines = []
for b in brands:
    g = green_flag.get(b, {})
    if not g:
        continue
    gw_lines.append(f"- **{b}**: green_share={_pct(g['green_share'])}, green_med={_fmt(g['green_median'])} vs overall_med={_fmt(g['overall_median'])} ‚Üí {'posible greenwashing' if g['possible_greenwashing'] else 'sin se√±al clara'}.")

# 7.5 p√°rrafo
para = (
    f"Yo apoyar√≠a como apuesta principal a **{pick_brand}** para crecer en redes en los pr√≥ximos 12 meses, "
    f"porque lidera en **{metric} mediano** (eficiencia por post) frente a la otra marca y eso suele traducirse en mejor retorno por d√≥lar de pauta. "
    f"Ajustar√≠a el mix subiendo el peso de **{_top_list(adjust_mix_up)}** (formatos con mejor performance) y recortando gradualmente los formatos m√°s frecuentes que no est√°n liderando el engagement "
    f"(**{_top_list(adjust_mix_down)}**, valid√°ndolo con pruebas A/B). "
    f"En calendario, concentrar√≠a los lanzamientos en la **ventana de oro {ab_str}** y har√≠a experimentos en ventanas desatendidas (poco volumen pero alto promedio) para ampliar alcance eficiente. "
    f"En KPIs, medir√≠a √©xito con **mediana** de engagement (no solo media), % de posts en el **cuartil superior** de {metric}, cobertura de m√©tricas (views presentes) y consistencia por franja (d√≠a/hora)."
)

md = "## 7. Preguntas de Negocio Finales (Entregable Ejecutivo)\n\n"

md += "### 7.1. Diferencia estrat√©gica (posicionamiento)\n"
md += "\n".join(s_lines) + "\n"
md += "- Lectura ejecutiva: una marca tiende a priorizar volumen/alcance si su mix est√° cargado a formatos de alta frecuencia pero con menor mediana; la otra prioriza eficiencia si concentra mejor mediana en categor√≠as espec√≠ficas y ventanas horarias consistentes.\n\n"

md += "### 7.2. Eficiencia vs volumen\n"
if winner_row is not None and loser_row is not None:
    md += f"- **Ganando atenci√≥n eficiente:** **{winner_eff}** (mediana {metric}={_fmt(winner_row['metric_median'])}) vs {loser_eff} (mediana={_fmt(loser_row['metric_median'])}).\n"
    md += f"- Volumen de posts: {winner_eff}={int(winner_row['posts'])} posts, {loser_eff}={int(loser_row['posts'])} posts.\n"
    md += f"- Si apuesto USD 1M en pauta+org√°nico para una sola marca: **{winner_eff}**, porque convertir views en interacci√≥n de forma consistente reduce el riesgo de depender de pocos virales y mejora el ROI esperado.\n\n"
else:
    md += "- No hay suficiente informaci√≥n para comparar eficiencia vs volumen.\n\n"

md += "### 7.3. Hora de oro (A/B conjunto)\n"
md += f"- Recomiendo ejecutar el A/B en **{ab_str}**, porque es una franja donde **ambas marcas** muestran desempe√±o alto (ajustado por su propio patr√≥n) con soporte suficiente de publicaciones, aumentando la probabilidad de √©xito y comparabilidad.\n\n"

md += "### 7.4. Greenwashing detector (opcional avanzado)\n"
if gw_lines:
    md += "\n".join(gw_lines) + "\n"
    md += "- Riesgo reputacional: si el p√∫blico percibe inconsistencia (discurso verde sin tracci√≥n o sin pruebas), puede generar desconfianza, comentarios negativos y p√©rdida de credibilidad de marca.\n\n"
else:
    md += "- No se detect√≥ se√±al clara o no se analiz√≥ contenido ‚Äúeco‚Äù.\n\n"

md += "### 7.5. Recomendaci√≥n ejecutiva (1 p√°rrafo)\n"
md += para + "\n"

print(md)
```
